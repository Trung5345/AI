```python
import math
import os

def GetWinner(board):
    if board[0] == board[1] == board[2] and board[0] != " ": return board[0]
    if board[3] == board[4] == board[5] and board[3] != " ": return board[3]
    if board[6] == board[7] == board[8] and board[6] != " ": return board[6]
    if board[0] == board[3] == board[6] and board[0] != " ": return board[0]
    if board[1] == board[4] == board[7] and board[1] != " ": return board[1]
    if board[2] == board[5] == board[8] and board[2] != " ": return board[2]
    if board[0] == board[4] == board[8] and board[0] != " ": return board[0]
    if board[2] == board[4] == board[6] and board[2] != " ": return board[2]
    return None

def PrintBoard(board):
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f"\n {board[0]} | {board[1]} | {board[2]}")
    print("-----------")
    print(f" {board[3]} | {board[4]} | {board[5]}")
    print("-----------")
    print(f" {board[6]} | {board[7]} | {board[8]}\n")

def GetAvailableCells(board):
    return [i + 1 for i, cell in enumerate(board) if cell != "X" and cell != "O"]

def alpha_beta(position, depth, alpha, beta, isMaximizing):
    winner = GetWinner(position)
    if winner == "X": return 10 - depth
    if winner == "O": return -10 + depth
    if len(GetAvailableCells(position)) == 0: return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            original_val = position[cell - 1]
            position[cell - 1] = "X"
            eval = alpha_beta(position, depth + 1, alpha, beta, False)
            position[cell - 1] = original_val
            maxEval = max(maxEval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha: break
        return maxEval
    else:
        minEval = math.inf
        for cell in GetAvailableCells(position):
            original_val = position[cell - 1]
            position[cell - 1] = "O"
            eval = alpha_beta(position, depth + 1, alpha, beta, True)
            position[cell - 1] = original_val
            minEval = min(minEval, eval)
            beta = min(beta, eval)
            if beta <= alpha: break
        return minEval

def FindBestMove(board, AI_player):
    bestVal = -math.inf if AI_player == "X" else math.inf
    bestMove = -1
    for cell in GetAvailableCells(board):
        original_val = board[cell - 1]
        board[cell - 1] = AI_player
        moveVal = alpha_beta(board, 0, -math.inf, math.inf, AI_player == "O")
        board[cell - 1] = original_val
        if AI_player == "X" and moveVal > bestVal:
            bestVal = moveVal
            bestMove = cell
        elif AI_player == "O" and moveVal < bestVal:
            bestVal = moveVal
            bestMove = cell
    return bestMove

def main():
    board = [str(i+1) for i in range(9)]
    player = input("Chọn X hoặc O: ").upper()
    AI = "O" if player == "X" else "X"
    current_turn = "X"

    while True:
        PrintBoard(board)
        if GetWinner(board):
            print(f"Người chơi {GetWinner(board)} chiến thắng!")
            break
        if len(GetAvailableCells(board)) == 0:
            print("Hòa!")
            break

        if current_turn == player:
            move = int(input("Nhập số ô (1-9): "))
            if board[move-1] not in ["X", "O"]:
                board[move-1] = player
                current_turn = AI
        else:
            print("Máy đang tính toán...")
            move = FindBestMove(board, AI)
            board[move-1] = AI
            current_turn = player

if __name__ == "__main__":
    main()
```

```python
import math
import copy

X = "X"
O = "O"
EMPTY = None

def initial_state(n):
    return [[EMPTY for _ in range(n)] for _ in range(n)]

def player(board):
    count = 0
    for row in board:
        for cell in row:
            if cell:
                count += 1
    return O if count % 2 != 0 else X

def actions(board):
    res = set()
    for i in range(len(board)):
        for j in range(len(board)):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    new_board = copy.deepcopy(board)
    i, j = action
    new_board[i][j] = player(board)
    return new_board

def winner(board):
    n = len(board)
    for i in range(n):
        if all(board[i][j] == board[i][0] != EMPTY for j in range(n)):
            return board[i][0]
        if all(board[j][i] == board[0][i] != EMPTY for j in range(n)):
            return board[0][i]
    if all(board[i][i] == board[0][0] != EMPTY for i in range(n)):
        return board[0][0]
    if all(board[i][n-1-i] == board[0][n-1] != EMPTY for i in range(n)):
        return board[0][n-1]
    return None

def terminal(board):
    if winner(board) is not None:
        return True
    return all(all(cell != EMPTY for cell in row) for row in board)

def heuristic(board):
    res = winner(board)
    if res == X: return 100
    if res == O: return -100
    return 0

def alpha_beta(board, depth, alpha, beta, is_maximizing):
    if terminal(board) or depth == 0:
        return heuristic(board)
    
    if is_maximizing:
        v = -math.inf
        for action in actions(board):
            v = max(v, alpha_beta(result(board, action), depth - 1, alpha, beta, False))
            alpha = max(alpha, v)
            if beta <= alpha:
                break
        return v
    else:
        v = math.inf
        for action in actions(board):
            v = min(v, alpha_beta(result(board, action), depth - 1, alpha, beta, True))
            beta = min(beta, v)
            if beta <= alpha:
                break
        return v

def get_best_move(board, n):
    curr_player = player(board)
    best_move = None
    depth_limit = 4 if n == 3 else 3
    
    if curr_player == X:
        v = -math.inf
        for action in actions(board):
            score = alpha_beta(result(board, action), depth_limit - 1, -math.inf, math.inf, False)
            if score > v:
                v = score
                best_move = action
    else:
        v = math.inf
        for action in actions(board):
            score = alpha_beta(result(board, action), depth_limit - 1, -math.inf, math.inf, True)
            if score < v:
                v = score
                best_move = action
    return best_move

if __name__ == "__main__":
    n = int(input("Kich thuoc N: "))
    board = initial_state(n)
    while not terminal(board):
        p = player(board)
        if p == X:
            r, c = map(int, input("Hang Cot: ").split())
            board[r][c] = X
        else:
            move = get_best_move(board, n)
            board[move[0]][move[1]] = O
        for row in board: print(row)
    print("Ket thuc:", winner(board))
```

```python
import math
import copy

PIECE_VALUES = {
    'P': 10, 'N': 30, 'B': 30, 'R': 50, 'Q': 90, 'K': 900,
    'p': -10, 'n': -30, 'b': -30, 'r': -50, 'q': -90, 'k': -900
}

def initial_chess_board():
    return [
        ['r', 'n', 'b', 'q', 'k', 'b', 'n', 'r'],
        ['p', 'p', 'p', 'p', 'p', 'p', 'p', 'p'],
        [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
        [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
        [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
        [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
        ['P', 'P', 'P', 'P', 'P', 'P', 'P', 'P'],
        ['R', 'N', 'B', 'Q', 'K', 'B', 'N', 'R']
    ]

def heuristic_evaluation(board):
    score = 0
    for row in board:
        for piece in row:
            if piece != ' ':
                score += PIECE_VALUES.get(piece, 0)
    return score

def get_all_moves(board, player_type):
    moves = []
    for r in range(8):
        for c in range(8):
            piece = board[r][c]
            if piece == ' ': continue
            if (player_type == "MAX" and piece.isupper()) or (player_type == "MIN" and piece.islower()):
                if r + (1 if player_type == "MIN" else -1) in range(8):
                    new_r = r + (1 if player_type == "MIN" else -1)
                    if board[new_r][c] == ' ':
                        moves.append(((r, c), (new_r, c)))
    return moves

def apply_move(board, move):
    new_board = copy.deepcopy(board)
    (start_r, start_c), (end_r, end_c) = move
    new_board[end_r][end_c] = new_board[start_r][start_c]
    new_board[start_r][start_c] = ' '
    return new_board

def alpha_beta(board, depth, alpha, beta, is_maximizing):
    if depth == 0:
        return heuristic_evaluation(board)

    if is_maximizing:
        max_eval = -math.inf
        for move in get_all_moves(board, "MAX"):
            evaluation = alpha_beta(apply_move(board, move), depth - 1, alpha, beta, False)
            max_eval = max(max_eval, evaluation)
            alpha = max(alpha, evaluation)
            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = math.inf
        for move in get_all_moves(board, "MIN"):
            evaluation = alpha_beta(apply_move(board, move), depth - 1, alpha, beta, True)
            min_eval = min(min_eval, evaluation)
            beta = min(beta, evaluation)
            if beta <= alpha:
                break
        return min_eval

def find_best_move(board, depth):
    best_move = None
    max_val = -math.inf
    alpha = -math.inf
    beta = math.inf
    
    for move in get_all_moves(board, "MAX"):
        val = alpha_beta(apply_move(board, move), depth - 1, alpha, beta, False)
        if val > max_val:
            max_val = val
            best_move = move
        alpha = max(alpha, val)
    return best_move

if __name__ == "__main__":
    board = initial_chess_board()
    print("AI Chess Alpha-Beta (Simplified Move Logic)")
    for row in board: print(row)
    
    print("\nAI dang tinh toan nuoc di tot nhat...")
    best = find_best_move(board, depth=3)
    if best:
        print(f"Nuoc di de xuat: {best[0]} -> {best[1]}")
        board = apply_move(board, best)
        for row in board: print(row)
    else:
        print("Khong tim thay nuoc di hop le.")
```
