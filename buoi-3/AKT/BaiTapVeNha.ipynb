```python
import random


def create_goal(n):
    goal = []
    num = 1
    for i in range(n):
        row = []
        for j in range(n):
            if num == n*n:
                row.append(0)
            else:
                row.append(num)
            num += 1
        goal.append(row)
    return goal



def to_tuple(state):
    return tuple(tuple(row) for row in state)



def heuristic(state, goal, n):
    dist = 0
    pos = {}

    for i in range(n):
        for j in range(n):
            pos[goal[i][j]] = (i, j)

    for i in range(n):
        for j in range(n):
            val = state[i][j]
            if val != 0:
                gx, gy = pos[val]
                dist += abs(i - gx) + abs(j - gy)

    return dist



def find_zero(state, n):
    for i in range(n):
        for j in range(n):
            if state[i][j] == 0:
                return i, j



def neighbors(state, n):
    x, y = find_zero(state, n)
    moves = [(-1,0),(1,0),(0,-1),(0,1)]
    result = []

    for dx, dy in moves:
        nx, ny = x + dx, y + dy
        if 0 <= nx < n and 0 <= ny < n:
            new_state = [row[:] for row in state]
            new_state[x][y], new_state[nx][ny] = new_state[nx][ny], new_state[x][y]
            result.append(new_state)

    return result



def AKT(start, n):
    goal = create_goal(n)

    OPEN = []
    CLOSED = set()

    OPEN.append({
        "state": start,
        "g": 0,
        "h": heuristic(start, goal, n),
        "parent": None
    })

    while OPEN:
        # chọn node có f nhỏ nhất
        min_f = min(node["g"] + node["h"] for node in OPEN)
        candidates = [node for node in OPEN if node["g"] + node["h"] == min_f]

        # chọn ngẫu nhiên (đúng AKT)
        current = random.choice(candidates)
        OPEN.remove(current)

        t = to_tuple(current["state"])
        if t in CLOSED:
            continue

        CLOSED.add(t)

        # nếu đạt đích
        if current["state"] == goal:
            path = []
            while current:
                path.append(current["state"])
                current = current["parent"]
            return path[::-1]

        # mở các node con
        for next_state in neighbors(current["state"], n):
            tt = to_tuple(next_state)

            if tt in CLOSED:
                continue

            if any(to_tuple(n["state"]) == tt for n in OPEN):
                continue

            OPEN.append({
                "state": next_state,
                "g": current["g"] + 1,
                "h": heuristic(next_state, goal, n),
                "parent": current
            })

    return None



n = 3   

start = [
 [1,2,3],
 [5,0,6],
 [4,7,8]
]

result = AKT(start, n)

if result:
    print("Số bước:", len(result)-1)
    for step in result:
        for row in step:
            print(row)
        print("------")
else:
    print("Không tìm thấy lời giải")
```
