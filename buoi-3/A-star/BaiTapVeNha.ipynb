```python
import heapq

def A_star_delivery(graph, start, heuristic):

    # trạng thái: (cost, current_node, visited_set, path)
    pq = []
    heapq.heappush(pq, (0, start, frozenset([start]), [start]))

    while pq:
        cost, current, visited, path = heapq.heappop(pq)

        # nếu đã đi hết tất cả node
        if len(visited) == len(graph):
            print("Đường đi:", path)
            print("Chi phí:", cost)
            return path

        for neighbor, weight in graph[current]:
            new_visited = visited | {neighbor}
            new_cost = cost + weight

            # f = g + h
            f = new_cost + heuristic[neighbor]

            heapq.heappush(pq, (f, neighbor, new_visited, path + [neighbor]))

    return None

graph = {
    'A': [('B',2), ('C',4)],
    'B': [('A',2), ('C',1), ('D',7)],
    'C': [('A',4), ('B',1), ('D',3)],
    'D': [('B',7), ('C',3)]
}

heuristic = {
    'A':3,
    'B':2,
    'C':1,
    'D':0
}

A_star_delivery(graph, 'A', heuristic)
```
