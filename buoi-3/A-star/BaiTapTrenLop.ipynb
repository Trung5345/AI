```python
def A_star(graph, start, stop, heuristic):

    open_set = set([start])
    closed_set = set([])

    g = {}
    g[start] = 0

    par = {}
    par[start] = start

    while len(open_set) > 0:
        n = None

        # chọn node có f nhỏ nhất
        for v in open_set:
            if n == None or g[v] + heuristic[v] < g[n] + heuristic[n]:
                n = v

        # không tìm thấy đường
        if n == None:
            print('Không tồn tại đường đi!')
            return None

        # nếu tới đích
        if n == stop:
            path = []

            while par[n] != n:
                path.append(n)
                n = par[n]

            path.append(start)
            path.reverse()

            print("Đường đi:", path)
            print("Chi phí:", g[stop])
            return path

        # duyệt node kề
        for (m, weight) in graph[n]:

            if m not in open_set and m not in closed_set:
                open_set.add(m)
                par[m] = n
                g[m] = g[n] + weight

            else:
                if g[m] > g[n] + weight:
                    g[m] = g[n] + weight
                    par[m] = n

                    if m in closed_set:
                        closed_set.remove(m)
                        open_set.add(m)

        open_set.remove(n)
        closed_set.add(n)

    print('Không tồn tại đường đi!')
    return None

graph = {
    'A': [('B',1), ('C',3)],
    'B': [('D',1), ('E',5)],
    'C': [('F',2)],
    'D': [('G',1)],
    'E': [('G',2)],
    'F': [('G',3)],
    'G': []
}

heuristic = {
    'A':5, 'B':4, 'C':4,
    'D':2, 'E':2, 'F':3,
    'G':0
}

# tìm đường từ A đến G
A_star(graph, 'A', 'G', heuristic)
```
