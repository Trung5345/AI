```python
import random


def tao_mang(m, n):
    return [[random.randint(1, 20) for _ in range(n)] for _ in range(m)]


def xuat_mang(a):
    for row in a:
        print(row)


def xuat_dong(a, k):
    print(f"Dòng {k}:", a[k])


def xuat_cot(a, k):
    print(f"Cột {k}:")
    for row in a:
        print(row[k], end=" ")
    print()


def dong_tong_max_duoi_45(a):
    max_sum = -1
    index = -1

    for i in range(len(a)):
        s = 0
        for x in a[i]:
            s += x

        if s < 45 and s > max_sum:
            max_sum = s
            index = i

    return index, max_sum


def cot_tich_min(a):
    m = len(a)
    n = len(a[0])

    min_prod = float('inf')
    index = -1

    for j in range(n):
        prod = 1
        for i in range(m):
            prod *= a[i][j]

        if prod < min_prod:
            min_prod = prod
            index = j

    return index, min_prod


def dong_chan_cot_le(a):
    print("Phần tử dòng chẵn, cột lẻ:")
    for i in range(len(a)):
        for j in range(len(a[0])):
            if i % 2 == 0 and j % 2 == 1:
                print(a[i][j], end=" ")
    print()


def tbc_chan_dong_le(a):
    tong = 0
    dem = 0

    for i in range(len(a)):
        if i % 2 == 1:
            for x in a[i]:
                if x % 2 == 0:
                    tong += x
                    dem += 1

    if dem == 0:
        return 0
    return tong / dem


def tbc_bien(a):
    m = len(a)
    n = len(a[0])

    tong = 0
    dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1

    return tong / dem



m = int(input("Nhập số dòng m: "))
n = int(input("Nhập số cột n: "))

a = tao_mang(m, n)

print("\nMảng a:")
xuat_mang(a)

k = int(input("\nNhập dòng k: "))
xuat_dong(a, k)

k = int(input("Nhập cột k: "))
xuat_cot(a, k)

idx, s = dong_tong_max_duoi_45(a)
print("\nDòng có tổng < 45 lớn nhất:", idx, "với tổng =", s)

idx, p = cot_tich_min(a)
print("Cột có tích nhỏ nhất:", idx, "với tích =", p)

dong_chan_cot_le(a)

print("TBC chẵn dòng lẻ:", tbc_chan_dong_le(a))
print("TBC biên:", tbc_bien(a))
```

```python
#Bai15

from datetime import datetime


class SinhVien:
    def __init__(self, ma_sv, ten, nam_sinh, diem_tb):
        self.ma_sv = ma_sv
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb


def nhap_danh_sach(n):
    ds = []
    for i in range(n):
        print(f"\nNhập sinh viên thứ {i+1}:")
        ma = input("Mã SV: ")
        ten = input("Tên: ")
        nam = int(input("Năm sinh: "))
        diem = float(input("Điểm TB: "))
        ds.append(SinhVien(ma, ten, nam, diem))
    return ds



def dem_len_lop(ds):
    return sum(1 for sv in ds if sv.diem_tb >= 5)



def xuat_20_tuoi(ds):
    nam_hien_tai = datetime.now().year
    print("\nSinh viên đủ 20 tuổi:")
    for sv in ds:
        if nam_hien_tai - sv.nam_sinh >= 20:
            print(sv.ma_sv, "-", sv.ten)



def dem_he_dh(ds):
    return sum(1 for sv in ds if sv.ma_sv[2:4] == "DH")



def dem_ten_lan(ds):
    return sum(1 for sv in ds if sv.ten.split()[-1] == "Lan")



def dem_ho_phan(ds):
    return sum(1 for sv in ds if sv.ten.split()[0] == "Phan")





if __name__ == "__main__":
    n = int(input("Nhập số sinh viên: "))
    ds = nhap_danh_sach(n)

    print("\n===== KẾT QUẢ =====")
    print("Số SV đủ điều kiện lên lớp:", dem_len_lop(ds))
    print("Số SV hệ đại học:", dem_he_dh(ds))
    print("Số SV tên Lan:", dem_ten_lan(ds))
    print("Số SV họ Phan:", dem_ho_phan(ds))

    xuat_20_tuoi(ds)
```
