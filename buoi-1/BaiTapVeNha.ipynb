```python
print("=== BÀI 13 ===")

a = list(map(int, input("Nhập mảng a : ").split()))
b = list(map(int, input("Nhập mảng b : ").split()))

print("Input a:", a)
print("Input b:", b)


def tron_mang(a, b):
    result = []
    n = min(len(a), len(b))

    for i in range(n):
        result.append(min(a[i], b[i]))

    if len(a) > n:
        result.extend(a[n:])
    else:
        result.extend(b[n:])

    return result


print("Mảng kết quả:", tron_mang(a, b))
```

```python
print("=== BÀI 2 ===")

a = [
    [1, -2, 3],
    [4, 5, -6],
    [7, 8, 9]
]

n = len(a)

print("Ma trận a:")
for row in a:
    print(row)


tong = 0
for i in range(n):
    for j in range(n):
        if i + j <= n - 1:
            tong += a[i][j]

print("\na. Tổng tam giác trên chéo phụ:", tong)


for i in range(n):
    for j in range(n):
        if a[i][j] < 0:
            a[i][j] = abs(a[i][j])

print("\nb. Sau khi đổi số âm:")
for row in a:
    print(row)


x = 100
for i in range(n):
    for j in range(n):
        if a[i][j] % 2 == 0:
            a[i][j] = x

print("\nc. Sau khi thay số chẵn bằng", x)
for row in a:
    print(row)


toan_chan = all(a[i][j] % 2 == 0 for i in range(n) for j in range(n))
print("\nd. Ma trận toàn chẵn?", toan_chan)


doi_xung = True
for i in range(n):
    for j in range(n):
        if a[i][j] != a[j][i]:
            doi_xung = False

print("\ne. Ma trận đối xứng?", doi_xung)


tang = True
for i in range(n - 1):
    if a[i][i] > a[i+1][i+1]:
        tang = False

print("\nf. Chéo chính tăng dần?", tang)


print("\ng. Tam giác dưới chéo phụ:")
for i in range(n):
    for j in range(n):
        if i + j >= n - 1:
            print(a[i][j], end=" ")
    print()


giam = True
cheo_phu = [a[i][n-1-i] for i in range(n)]

for i in range(len(cheo_phu) - 1):
    if cheo_phu[i] < cheo_phu[i+1]:
        giam = False

print("\nh. Chéo phụ giảm dần?", giam)
```
